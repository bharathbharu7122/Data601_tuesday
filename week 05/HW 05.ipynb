```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

## import numpy
import numpy as np

url = 'https://raw.githubusercontent.com/msaricaumbc/DS_data/master/nyc_flights.csv'
flight_1 = pd.read_csv(url)
```

```python
flight_1
```

```python
flight_1.info()
```

```python
flight_1.describe()
```

```python
plot = sns.catplot(data= flight_1, kind = 'count', x = 'carrier')
plot.ax.set_title('flight vs carrier ')
```

```python
plot = sns.displot(data = flight_1, x= 'origin', kind = 'hist', height=4,aspect=1)
plot.ax.grid(axis = 'x')
```

```python
plot = sns.displot(data = flight_1, x= 'flight', kind = 'kde', height= 3, aspect = 2)
plot.ax.grid(axis = 'x')
```

```python
plot = sns.displot(data = flight_1, x= 'month', kind = 'ecdf', height= 4, aspect = 3)
plot.ax.grid(axis = 'x')
```

```python
plot = sns.displot(flight_1, x="carrier", y="day", cbar = True)
plot.ax.tick_params(labelrotation = 90)
```

```python
plot= sns.displot(data=flight_1, kind="kde", x="flight", hue = 'day', aspect=1, alpha=1)
plot.ax.grid()
```

```python
sns.set(rc={"figure.figsize":(10, 8)}) 
sns.set_style('ticks')

plot = sns.boxplot(y = 'carrier', x = 'month', data = flight_1 )
```

```python
plt.scatter(data = flight_1, x="carrier", y="distance")
```

```python
x = flight_1.carrier
y = flight_1.dep_time
h = flight_1.distance
plot = sns.scatterplot(x= x, y = y,hue=h, alpha = 0.1 )
```
