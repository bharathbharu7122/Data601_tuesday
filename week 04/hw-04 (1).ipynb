```python
import pandas

dframe = pandas.read_csv(r"C:\Users\bhara\Downloads\RollingSystemDemand_20220227_1500.csv",
                         index_col=False,
                         skiprows=1,
                         skipfooter=1,
                         engine='python',
                         header=None)

dframe.columns=['VD','time of measurement','value']

dframe.head()
```

```python
def hourly_function(a):
    print (a.hour)
    return a.hour

dframe['time'] = pandas.to_datetime(dframe['time of measurement'].astype('str'))

dframe['time_hour'] = dframe['time_hour'].dt.hour
dframe['time_date'] 
```

```python
import matplotlib.pyplot as plt
plt.bar(dframe['time_hour'],dframe['value'])
plt.xlabel('Number of Hours', fontsize=12)
plt.ylabel('Value', fontsize=8)
plt.title('Sum of Value over hours', fontsize=12)
```

```python
fig_size = plt.rcParams["figure.figsize"]
print(fig_size)
fig_size[0] = 20
fig_size[1] = 100
plt.rcParams["figure.figsize"] = fig_size
dframe['value'].groupby([dframe['time_date'],dframe['time_hour']]).mean().plot(kind='barh')
plt.xlabel('Number of Hours', fontsize=4)
plt.ylabel('Value', fontsize=4)
plt.title('Values over hours', fontsize=4)
```
