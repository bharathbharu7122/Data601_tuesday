### Average Increase

```python
### Average Increase

## Average of all year-on-year percentage increments fom '22 to '27 

# Incremental percentage = (((new)-(old))/(old))*100
```

```python
df_comb['2022']
```

```python
'3,400'.replace(',','')
```

```python
df_comb['2022'] = pd.to_numeric(df_comb['2022'].str.replace(',',''))
df_comb['2023'] = pd.to_numeric(df_comb['2023'].str.replace(',',''))
df_comb['2024'] = pd.to_numeric(df_comb['2024'].str.replace(',',''))
df_comb['2025'] = pd.to_numeric(df_comb['2025'].str.replace(',',''))
df_comb['2026'] = pd.to_numeric(df_comb['2026'].str.replace(',',''))
df_comb['2027'] = pd.to_numeric(df_comb['2027'].str.replace(',',''))
```

```python
df_comb['inc_22_23'] = (((df_comb['2023'])-(df_comb['2022']))/(df_comb['2022']))*100
df_comb.head()
```

```python
df_comb['inc_23_24'] = (((df_comb['2024'])-(df_comb['2023']))/(df_comb['2023']))*100
df_comb['inc_24_25'] = (((df_comb['2025'])-(df_comb['2024']))/(df_comb['2024']))*100
df_comb['inc_25_26'] = (((df_comb['2026'])-(df_comb['2025']))/(df_comb['2025']))*100
df_comb['inc_26_27'] = (((df_comb['2027'])-(df_comb['2026']))/(df_comb['2026']))*100
```

```python
df_comb.head()
```

```python
df_comb.loc[:,'inc_22_23':'inc_26_27'].head()
```

```python
df_comb['Avg Increase %'] = \
(df_comb['inc_22_23']+df_comb['inc_23_24']+df_comb['inc_24_25']+df_comb['inc_25_26']+df_comb['inc_26_27'])/5
```

```python
df_comb.head()
```

```python
# removing some columns

del df_comb['inc_22_23']
del df_comb['inc_23_24']
del df_comb['inc_24_25']
del df_comb['inc_25_26']
del df_comb['inc_26_27']
```

```python
df_comb.head()
```

```python
## Rounding it to three decimals 
```

```python
df_comb['Avg Increase %'] = df_comb['Avg Increase %'].map(lambda a: round(a,3))
```

```python
df_comb.head()
```

### Sorting by Avg Increase % in descending order.

```python
df_comb = df_comb.sort_values(by='Avg Increase %', ascending=False).head()
```

```python
df_comb.reset_index(inplace=True)
```
