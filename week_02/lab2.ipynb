* Height only

```python
import pandas as pd

nba_players = pd.read_csv('https://raw.githubusercontent.com/msaricaumbc/DS_data/master/all_seasons.csv', index_col = 0)

height = nba_players['player_height']
points = nba_players['pts']
```

```python
import matplotlib.pyplot as plt
#df = pd.DataFrame([{'Player_heights' : height}])
#_=plt.boxplot(height)
#_=plt.scatter(height,points)
list_height=[]
height_values=[]

for i in range(len(height)):
    list_height.append(height[i])
    
for i in range(len(height)):
    height_values.append(i)
   
```

```python
#plot's for player's height's

#Scatter plot for height of the players
plt.title("Height_visualization")
plt.scatter(height_values,list_height)
plt.ylabel('Height_of_the_palyers')
plt.xlabel("Player_Number")
plt.show()

#Histogram for height of the player's

plt.title("Height_visualization")
plt.hist(list_height)
plt.ylabel("Number_of_players")
plt.xlabel("Player's_Height")
plt.grid(True)
plt.show()
```

```python
points_list = []
player_number_list = []

for j in range(len(points)):
    points_list.append(points[j])
for j in range(len(points)):
    player_number_list.append(j)

#plot for point's

plt.title("Point's_visualization")
plt.ylabel("Point's_Scored")
plt.xlabel("Player's_Number")
plt.bar(player_number_list,points_list)
plt.show()
```

```python
#plot for height vs  points

height_list=[]
for i in range(len(height)):
    height_list.append(i)
plt.scatter(list_height,points_list)
plt.title("Player's_height vs Points")
plt.xlabel('Player_height')
plt.ylabel('Points')
plt.draw()
```

## Lab Part2
Please visualize year vs score with correct/appropriate labels

```python
bowling_data = pd.read_csv("https://raw.githubusercontent.com/msaricaumbc/DS_data/master/bowling_stats.csv",
                            header=None,
                            names=['year','city','state','count1','count2'])
def merge_columns(row):
    if pd.isna(row['count2']):
        return row['count1']
    else:
        return row['count1']*1000+row['count2']
    
bowling_data['total']=bowling_data.apply(merge_columns,axis=1)

year = bowling_data['year']
score = bowling_data['total']

years=[]
scores=[]
for l in range(len(year)):
    years.append(year[l])
for m in range(len(score)):
    scores.append(score[m])

#plot for year vs score

plt.title('Year vs Score')
plt.xlabel("Year's")
plt.ylabel("Score's")
plt.scatter(years,scores)
plt.show()
```
