```python
import pandas as pd
```

```python
spicy = 'https://raw.githubusercontent.com/msaricaumbc/DS_Data/master/strings/hw_data_spicy.csv'
less_spicy = 'https://raw.githubusercontent.com/msaricaumbc/DS_Data/master/strings/hw_data_less_spicy.csv'
```

```python
df = pd.read_csv(spicy, index_col=0)
```

```python
df.head()
```

```python
import re
start_column =[];
end_column =[];
phonenumber = [];
for i in df['notes']:
  d1 = re.sub('end dt|end date|ed date|endt',"end date", i)
  d2 = re.sub('start date|STDT|st date|stdt', 'start date', d1)
  d3 = re.sub('P#|pn|p#|p/n', 'Phone Number', d2)
  index=d3.find('end date')
  index2=d3.find('start date')
  index4=d3.find('Phone Number')
  if('Phone Number' in d3):
    if(d3[0] =='P'):
      phonenumber.append(d3[13:index2-1].strip())
    else:
      phonenumber.append(d3[index4+13:].strip())
  else:
    phonenumber.append('None')
  d4 = d3[index2:index-1]
  d5 = d3[index:]
  if('Phone Number' in d5):
    d5 = d5
  else:
    d5 = (d5+ ' ' +'Phone Number')
  index3 = d5.find('Phone Number')
  end_date = d5[9:index3-1]
  start_date = d4[d4.find('start date') + 10:].replace(":","").strip()
  try:
    start_column.append(pd.to_datetime(start_date))
  except Exception:
    start_column.append(pd.to_datetime(start_date[:len(start_date)-2] + str(int(start_date[len(start_date)-2:]) +2000)))
  try:
    end_column.append(pd.to_datetime(end_date))
  except Exception:
    end_column.append(pd.to_datetime(end_date[:len(end_date)-2] + str(int(end_date[len(end_date)-2:]) +2000)))

# print(start_column)
# print(end_column)
# print(phonenumber)
```

```python
Question1 = pd.DataFrame({'Start_Date': start_column, 'End_Date': end_column})
Question1
```

```python
Question2 = pd.DataFrame({'Phone Number': phonenumber})
Question2
```
